```python
# –º–µ—Ç–æ–¥, –≤–æ–∑–≤—Ä–∞—â–∞—é—â–∏–π True, –µ—Å–ª–∏ –¥–≤–∞ –æ—Ç—Ä–µ–∑–∫–∞ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è. –ù–µ–æ–±—Ö–æ–¥–∏–º –¥–ª—è —Ç–æ–≥–æ,
# —á—Ç–æ–±—ã –ø—Ä–æ–≤–µ—Ä—è—Ç—å –Ω–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç—å –≤–≤–æ–¥–∏–º—ã–µ –¥–∞–Ω–Ω—ã–µ
def intersects(p1: tuple[int, int], p2: tuple[int, int], p3: tuple[int, int], p4: tuple[int, int]) -> bool:
    common = (p2[0] - p1[0])*(p4[1] - p3[1]) - (p2[1] - p1[1])*(p4[0] - p3[0])

    if common == 0:
        return False

    r_h = (p1[1] - p3[1]) * (p4[0] - p3[0]) - (p1[0] - p3[0]) * (p4[1] - p3[1])
    s_h = (p1[1] - p3[1]) * (p2[0] - p1[0]) - (p1[0] - p3[0]) * (p2[1] - p1[1])

    r = r_h / common
    s = s_h / common

    if 0 <= r <= 1 and 0 <= s <= 1:
        return True
    else:
        return False


def danger_turns(k: int, input_func) -> int:
    if not 4 <= k <= 1000:
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 4 ‚â§ k ‚â§ 1000')

    passed_points = []

    def validate_input():
        x, y = input_func()

        if len(passed_points) == 1 and (x - passed_points[0][0] != 0 or y - passed_points[0][1] <= 0):
            raise ValueError('–ø–µ—Ä–≤—ã–π –ø—Ä—è–º–æ–π —É—á–∞—Å—Ç–æ–∫ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω –Ω–∞ —Å–µ–≤–µ—Ä')

        if (x, y) in passed_points and len(passed_points) != k:
            raise ValueError('–Ω–∏–∫–∞–∫–∞—è –ø–∞—Ä–∞ —Ç–æ—á–µ–∫ (–∫—Ä–æ–º–µ –ø–µ—Ä–≤–æ–π –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–π) –Ω–µ –º–æ–∂–µ—Ç —Å–æ–≤–ø–∞–¥–∞—Ç—å')

        if len(passed_points) == k and (x, y) != passed_points[0]:
            raise ValueError('–ø–æ—Å–ª–µ–¥–Ω—è—è —Ç–æ—á–∫–∞ –¥–æ–ª–∂–Ω–∞ —Å–æ–≤–ø–∞–¥–∞—Ç—å —Å –ø–µ—Ä–≤–æ–π')

        if len(passed_points) > 0:
            if y < passed_points[0][1] or (y == passed_points[0][1] and x < passed_points[0][0]):
                raise ValueError('—Å–∞–º–∞—è —é–∂–Ω–∞—è –∏ –∑–∞–ø–∞–¥–Ω–∞—è —Ç–æ—á–∫–∞ —Ç—Ä–∞—Å—Å—ã –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø–µ—Ä–≤–æ–π —Ç–æ—á–∫–æ–π')
                
            if x - passed_points[-1][0] != 0 and y - passed_points[-1][1] != 0:
                raise ValueError('–¥–≤–∏–∂–µ–Ω–∏–µ –º–æ–∂–µ—Ç –±—ã—Ç—å —Ç–æ–ª—å–∫–æ –Ω–∞ —Å–µ–≤–µ—Ä, —é–≥, –∑–∞–ø–∞–¥ –∏–ª–∏ –≤–æ—Å—Ç–æ–∫')

        for num in [x, y]:
            if not -10000 <= num <= 10000:
                raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ -10000 ‚â§ xi,yi ‚â§ 10000')

        for i in range(0 if len(passed_points) != k else 1, len(passed_points) - 2):
            if intersects(passed_points[i], passed_points[i + 1], passed_points[-1], (x, y)):
                raise ValueError('–ª—é–±–∞—è –ø–∞—Ä–∞ –ø—Ä—è–º—ã—Ö —É—á–∞—Å—Ç–∫–æ–≤ —Ç—Ä–∞—Å—Å—ã –Ω–µ –¥–æ–ª–∂–Ω–∞ –∏–º–µ—Ç—å –æ–±—â–∏—Ö —Ç–æ—á–µ–∫')

        passed_points.append((x, y))
        return x, y

    ax, ay = validate_input()
    bx, by = validate_input()
    cx, cy = validate_input()

    result = 0

    for i in range(k - 2):
        vec_a_x = bx - ax
        vec_a_y = by - ay

        vec_b_x = cx - bx
        vec_b_y = cy - by

        z = vec_a_x * vec_b_y - vec_a_y * vec_b_x

        if z == 0:
            raise ValueError('–¥–≤–∞ —Å–æ—Å–µ–¥–Ω–∏—Ö —É—á–∞—Å—Ç–∫–∞ –Ω–µ –º–æ–≥—É—Ç –±—ã—Ç—å –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω—ã –≤ –æ–¥–Ω—É –∏ —Ç—É –∂–µ —Å—Ç–æ—Ä–æ–Ω—É –∏–ª–∏ –≤ '
                             '–ø—Ä–æ—Ç–∏–≤–æ–ø–æ–ª–æ–∂–Ω—ã–µ —Å—Ç–æ—Ä–æ–Ω—ã')
        elif z > 0:
            result += 1

        ax, ay = bx, by
        bx, by = cx, cy
        cx, cy = validate_input()

    return result


print('–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: ')
k = int(input())
result = danger_turns(k, lambda: map(int, input().split()))

print('\n–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: ')
print(result)
```

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–µ–¥–∏—Ç–µ –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –≤–æ–∑–º–æ–∂–Ω—É—é —Å—É–º–º–∞—Ä–Ω—É—é —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–¥–µ—Ä–∂–∫–∏ –≤—Å–µ—Ö —Ä–µ–π—Å–æ–≤.

–ü—Ä–∏–º–µ—Ä\

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ\
–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –ø—Ä–∏–º–µ—Ä –∏–∑ —É—Å–ª–æ–≤–∏—è. –ï—Å–ª–∏ —Å–¥–≤–∏–Ω—É—Ç—å —Ä–µ–π—Å—ã –Ω–∞ 2 –º–∏–Ω—É—Ç—ã, —Å–æ—Ö—Ä–∞–Ω–∏–≤ –∏—Ö –ø–æ—Ä—è–¥–æ–∫ –≤—ã–ª–µ—Ç–∞, —Ç–æ —Å—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–¥–µ—Ä–∂–∫–∏ –≤—Å–µ—Ö —Ä–µ–π—Å–æ–≤ —Å–æ—Å—Ç–∞–≤–∏—Ç (3‚Äâ-‚Äâ1)¬∑4‚Äâ+‚Äâ(4‚Äâ-‚Äâ2)¬∑2‚Äâ+‚Äâ(5‚Äâ-‚Äâ3)¬∑1‚Äâ+‚Äâ(6‚Äâ-‚Äâ4)¬∑10‚Äâ+‚Äâ(7‚Äâ-‚Äâ5)¬∑2‚Äâ=‚Äâ38 –±—É—Ä–ª–µ–π.

–ï—Å–ª–∏ –∂–µ –∏–∑–º–µ–Ω–∏—Ç—å —Ä–∞—Å–ø–∏—Å–∞–Ω–∏–µ —Ç–∞–∫, –∫–∞–∫ –ø–æ–∫–∞–∑–∞–Ω–æ –≤ –æ—Ç–≤–µ—Ç–µ, —Ç–æ —Å—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å —Å–æ—Å—Ç–∞–≤–∏—Ç (3‚Äâ-‚Äâ1)¬∑4‚Äâ+‚Äâ(6‚Äâ-‚Äâ2)¬∑2‚Äâ+‚Äâ(7‚Äâ-‚Äâ3)¬∑1‚Äâ+‚Äâ(4‚Äâ-‚Äâ4)¬∑10‚Äâ+‚Äâ(5‚Äâ-‚Äâ5)¬∑2‚Äâ=‚Äâ20 –±—É—Ä–ª–µ–π.

```python
def cheapest(delay: int, costs: list[int]):
    if not 1 <= delay <= len(costs) <= 300000:
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 1 ‚â§ delay ‚â§ –¥–ª–∏–Ω–∞ costs ‚â§ 300000')

    for cost in costs:
        if not 1 <= cost <= 10 ** 7:
            raise ValueError(f'–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 1 ‚â§ {cost} (—Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–¥–µ—Ä–∂–∫–∏ —Ä–µ–π—Å–∞) ‚â§ 10^7')

    sorted_costs = []

    for i in range(len(costs)):
        index = max(0, i - delay)

        if len(sorted_costs) == 0:
            sorted_costs.append((costs[i], i))
            continue

        while index < len(sorted_costs) and costs[i] < sorted_costs[index][0]:
            index += 1

        sorted_costs.insert(index, (costs[i], i))

    price = 0
    best_order = []

    for i in range(len(sorted_costs)):
        flight = sorted_costs[i]
        price += flight[0] * (abs(flight[1] - (i + delay)))
        best_order.append(flight[1] + delay + 1)

    return price, best_order


assert cheapest(2, [4, 2, 1, 10, 2]) == (20, [3, 6, 7, 4, 5])
assert cheapest(2, [3, 1, 2]) == (11, [3, 5, 4])
assert cheapest(5, [5, 5, 9, 100, 3]) == (321, [9, 8, 7, 6, 10])
assert cheapest(1, [1]) == (1, [2])
assert cheapest(1, [10000000]) == (10000000, [2])

print('–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: ')
delay = int(input())
costs = [int(cost) for cost in input().split()]

result = cheapest(delay, costs)

print('\n–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: ')
print(result[0])
print(result[1])
```

## –ó–∞–¥–∞—á–∞ 4
–£ –ò–≤–∞–Ω–∞ –µ—Å—Ç—å –º–∞—Å—Å–∏–≤, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ n —ç–ª–µ–º–µ–Ω—Ç–æ–≤. –ö–∞–∂–¥—ã–π –∏–∑ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ ‚Äî —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ –æ—Ç 1 –¥–æ n.

–°–æ–≤—Å–µ–º –Ω–µ–¥–∞–≤–Ω–æ –ò–≤–∞–Ω —É–∑–Ω–∞–ª –æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞—Ö –∏ –∏—Ö –ª–µ–∫—Å–∏–∫–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–º –ø–æ—Ä—è–¥–∫–µ. –¢–µ–ø–µ—Ä—å –æ–Ω —Ö–æ—á–µ—Ç –∏–∑–º–µ–Ω–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ —Å–≤–æ—ë–º –º–∞—Å—Å–∏–≤–µ —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —á—Ç–æ–±—ã –µ–≥–æ –º–∞—Å—Å–∏–≤ —Å—Ç–∞–ª –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–æ–π (—Ç–æ –µ—Å—Ç—å –∫–∞–∂–¥–æ–µ –∏–∑ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ n –≤—Å—Ç—Ä–µ—á–∞–ª–æ—Å—å –≤ –µ–≥–æ –º–∞—Å—Å–∏–≤–µ —Ä–æ–≤–Ω–æ –ø–æ –æ–¥–Ω–æ–º—É —Ä–∞–∑—É). –ï—Å–ª–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –º–Ω–æ–≥–æ —Å–ø–æ—Å–æ–±–æ–≤ —Å–¥–µ–ª–∞—Ç—å –∑–∞–¥–∞–Ω–Ω—ã–π –º–∞—Å—Å–∏–≤ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–æ–π –∑–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑–º–µ–Ω–µ–Ω–∏–π, —Ç–æ —Å—Ä–µ–¥–∏ –Ω–∏—Ö –ò–≤–∞–Ω —Ö–æ—á–µ—Ç –≤—ã–±—Ä–∞—Ç—å —Ç–∞–∫–æ–π, —á—Ç–æ –ø–æ–ª—É—á–µ–Ω–Ω–∞—è –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –ª–µ–∫—Å–∏–∫–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–∞.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –≤ –ø–µ—Ä–≤—É—é –æ—á–µ—Ä–µ–¥—å –ò–≤–∞–Ω —Ö–æ—á–µ—Ç –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑–º–µ–Ω–µ–Ω–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –∞ –≤–æ –≤—Ç–æ—Ä—É—é ‚Äî –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –ª–µ–∫—Å–∏–∫–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –ø–æ–ª—É—á–µ–Ω–Ω—É—é –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫—É.

–î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –∫–∞–∫–∞—è –∏–∑ –¥–≤—É—Ö –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ –ª–µ–∫—Å–∏–∫–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –º–µ–Ω—å—à–µ, –æ–Ω–∏ —Å–Ω–∞—á–∞–ª–∞ —Å—Ä–∞–≤–Ω–∏–≤–∞—é—Ç—Å—è –ø–æ –ø–µ—Ä–≤–æ–º—É —ç–ª–µ–º–µ–Ω—Ç—É. –ü—Ä–∏ –∏—Ö —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ ‚Äî –ø–æ –≤—Ç–æ—Ä–æ–º—É, –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ. –°—Ä–µ–¥–∏ –¥–≤—É—Ö –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ x –∏ y –ª–µ–∫—Å–∏–∫–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –º–µ–Ω—å—à–µ –±—É–¥–µ—Ç x, –µ—Å–ª–∏ xi‚Äâ<‚Äâyi, –≥–¥–µ i ‚Äî –ø–µ—Ä–≤—ã–π –∏–Ω–¥–µ–∫—Å, –≤ –∫–æ—Ç–æ—Ä–æ–º –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ x –∏ y —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è.

–û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –∫–∞–∫ –±—É–¥–µ—Ç –≤—ã–≥–ª—è–¥–µ—Ç—å –º–∞—Å—Å–∏–≤ –ò–≤–∞–Ω–∞, –µ—Å–ª–∏ –æ–Ω –æ—Å—É—â–µ—Å—Ç–≤–∏—Ç –∑–∞–¥—É–º–∞–Ω–Ω–æ–µ.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ —Å–ª–µ–¥—É–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ n (2‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ200‚Äâ000) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –º–∞—Å—Å–∏–≤–µ –ò–≤–∞–Ω–∞.

–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ —Å–ª–µ–¥—É–µ—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª a1,‚Äâa2,‚Äâ...,‚Äâan (1‚Äâ‚â§‚Äâai‚Äâ‚â§‚Äân) ‚Äî –æ–ø–∏—Å–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ –ò–≤–∞–Ω–∞.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤—É—é —Å—Ç—Ä–æ–∫—É –≤—ã–≤–µ–¥–∏—Ç–µ q ‚Äî –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –Ω—É–∂–Ω–æ –∏–∑–º–µ–Ω–∏—Ç—å –≤ –º–∞—Å—Å–∏–≤–µ –ò–≤–∞–Ω–∞ –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –º–∞—Å—Å–∏–≤ —Å—Ç–∞–ª –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–æ–π. –í–æ –≤—Ç–æ—Ä—É—é —Å—Ç—Ä–æ–∫—É –≤—ã–≤–µ–¥–∏—Ç–µ –ª–µ–∫—Å–∏–∫–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–ª—É—á–∞–µ—Ç—Å—è —Å –ø–æ–º–æ—â—å—é q –∏–∑–º–µ–Ω–µ–Ω–∏–π.

–ü—Ä–∏–º–µ—Ä—ã\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
4\
3 2 2 3\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
2\
1 2 4 3 

```python
def permutation(nums: list[int]) -> tuple[int, list[int]]:
    if not 2 <= len(nums) <= 200000:
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 2 ‚â§ –¥–ª–∏–Ω–∞ nums ‚â§ 200000')

    for num in nums:
        if not 1 <= num <= len(nums):
            raise ValueError(f'–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 1 ‚â§ {num} (—á–∏—Å–ª–æ –≤ –º–∞—Å—Å–∏–≤–µ –ò–≤–∞–Ω–∞) ‚â§ {len(nums)} (–¥–ª–∏–Ω–∞ nums)')

    positions = dict()
    result = []
    count = 0

    # –∑–∞–ø–æ–ª–Ω—è–µ–º —Ç–∞–±–ª–∏—Ü—É: –≥–¥–µ –∫–ª—é—á - —á–∏—Å–ª–æ, –∑–Ω–∞—á–µ–Ω–∏–µ - —Å–ø–∏—Å–æ–∫ –∏–Ω–¥–µ–∫—Å–æ–≤, –≥–¥–µ —ç—Ç–æ —á–∏—Å–ª–æ —Å—Ç–æ–∏—Ç
    for i in range(len(nums)):
        num = nums[i]

        if num in positions:
            positions[num] += [i]
        else:
            positions[num] = [i]

    for i in range(len(nums)):
        num = nums[i]

        # –µ—Å–ª–∏ —á–∏—Å–ª–æ –Ω–∞ "—Å–≤–æ—ë–º" –º–µ—Å—Ç–µ –∏–ª–∏ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Ä–∞–∑
        if num == i + 1 or len(positions[num]) == 1:
            result.append(num)
            continue

        # –∏–Ω–∞—á–µ –Ω–∞—Ö–æ–¥–∏–º –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –∫–ª—é—á, –∫–æ—Ç–æ—Ä–æ–≥–æ –≤ —Ç–∞–±–ª–∏—Ü–µ –µ—â–µ –Ω–µ—Ç
        for j in range(1, len(nums) + 1):
            if j > num and positions[num][0] == i:
                result.append(num)
                break

            if j not in positions:
                # —Å—á–∏—Ç–∞–µ–º –∫–æ–ª-–≤–æ –∑–∞–º–µ–Ω, –¥–æ–±–∞–≤–ª—è–µ–º –≤ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∏–π —Å–ø–∏—Å–æ–∫,
                # —É–¥–∞–ª—è–µ–º –∏–∑ —Ç–∞–±–ª–∏—Ü—ã –∑–∞–º–µ–Ω—ë–Ω–Ω—ã–π –∏–Ω–¥–µ–∫—Å –∏ –¥–æ–±–∞–≤–ª—è–µ–º –≤ —Ç–∞–±–ª–∏—Ü—É –∏–Ω–¥–µ–∫—Å –Ω–æ–≤–æ–≥–æ —á–∏—Å–ª–∞
                count += 1
                result.append(j)
                positions[num].remove(i)
                positions[j] = [i]
                break

    return count, result


assert permutation([2, 2, 4, 4]) == (2, [1, 2, 3, 4])
assert permutation([2, 2, 3, 3]) == (2, [1, 2, 3, 4])
assert permutation([3, 2, 2, 2, 3]) == (3, [1, 2, 4, 5, 3])
assert permutation([1, 2, 2, 2, 3]) == (2, [1, 2, 4, 5, 3])
assert permutation([6, 8, 4, 6, 7, 1, 6, 3, 4, 5]) == (3, [2, 8, 4, 6, 7, 1, 9, 3, 10, 5])

print('–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
nums = [int(num) for num in input().split()]

result = permutation(nums)

print('\n–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
print(result[0])
print(result[1])
```

–°—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å 3+6+12+24+3+48+75=171. –ù–µ–ª—å–∑—è –ø–æ–ª—É—á–∏—Ç—å –º–µ–Ω—å—à—É—é —Å—Ç–æ–∏–º–æ—Å—Ç—å.

```python
import math


# a - —Å—Ç–æ–∏–º–æ—Å—Ç—å –ø–µ—Ä–µ–º–µ—â–µ–Ω–∏—è
# b - —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞—Ö–≤–∞—Ç–∞
def min_capture_cost(a: int, b: int, kingdoms: list[int]) -> int:
    if not 1 <= len(kingdoms) <= 2 * (10 ** 5):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 1 ‚â§ –¥–ª–∏–Ω–∞ kingdoms ‚â§ 2 * 10^5')

    if not 1 <= a <= 10 ** 5:
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 1 ‚â§ a ‚â§ 10^5')

    if not 1 <= b <= 10 ** 5:
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 1 ‚â§ b ‚â§ 10^5')

    for i in range(len(kingdoms)):
        prev = kingdoms[i - 1] if i > 0 else 0
        next = kingdoms[i + 1] if i < len(kingdoms) - 1 else 10 ** 8 + 1
        if not prev < kingdoms[i] < next:
            raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 1 ‚â§ x1 < x2 < ‚Ä¶ < xn ‚â§ 10^8')

    if a <= b:
        current_kingdom = 0
        min_cost = 0

        for kingdom in kingdoms[:-1]:
            min_cost += b * (kingdom - current_kingdom)
            min_cost += a * (kingdom - current_kingdom)
            current_kingdom = kingdom

        min_cost += b * (kingdoms[-1] - current_kingdom)

        return min_cost
    else:
        min_cost = math.inf

        for i in range(len(kingdoms)):
            temp_min = 0
            current_kingdom = 0
            remaining_moves = i

            for kingdom in kingdoms:
                temp_min += b * (kingdom - current_kingdom)

                if remaining_moves > 0:
                    remaining_moves -= 1
                    temp_min += a * (kingdom - current_kingdom)
                    current_kingdom = kingdom

            if temp_min < min_cost:
                min_cost = temp_min
            else:
                break

        return min_cost


assert min_capture_cost(2, 7, [3, 5, 12, 13, 21]) == 173
assert min_capture_cost(6, 3, [1, 5, 6, 21, 30]) == 171
assert min_capture_cost(9, 3, [10, 15]) == 75
assert min_capture_cost(27182, 31415, [16, 18, 33, 98, 874, 989, 4848, 20458, 34365, 38117, 72030]) == 3298918744

print('–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
t = int(input())
results = []
n_sum = 0

for i in range(t):
    a, b = map(int, input().split())
    kingdoms = list(map(int, input().split()))
    n_sum += len(kingdoms)

    if n_sum > 2 * (10 ** 5):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ: —Å—É–º–º–∞ –¥–ª–∏–Ω–Ω –Ω–∞–±–æ—Ä–æ–≤ –∫–æ—Ä–æ–ª–µ–≤—Å—Ç–≤ –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 2‚ãÖ10^5')

    results.append(min_capture_cost(a, b, kingdoms))

print('\n–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
print(results)
```

```python
# –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ —Å–ø–∏—Å–∫—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∏ '—É–±–∏—Ä–∞–µ–º' –æ–¥–Ω—É –∑–∞–∫—Ä—ã–≤–∞—é—â—É—é—Å—è
# —Å–∫–æ–±–∫—É, –µ—Å–ª–∏ —ç—Ç–æ –≤–æ–∑–º–æ–∂–Ω–æ. –≠—Ç–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã —É–±—Ä–∞–Ω–Ω—É—é
# —Å–∫–æ–±–∫—É –ø–æ—Å—Ç–∞–≤–∏—Ç—å –≤–º–µ—Å—Ç–æ –Ω–æ–≤–æ–π –Ω–∞–π–¥–µ–Ω–Ω–æ–π —Ä–µ—à—ë—Ç–∫–∏
def grab_closing(result: list[int]) -> bool:
    for i in range(len(result) - 1, -1, -1):
        if result[i] > 1:
            result[i] -= 1
            return True

    return False


def beautify_count(src: str) -> list[int] or int:
    if not 1 <= len(src) <= 10 ** 5:
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 1 ‚â§ –¥–ª–∏–Ω–∞ src ‚â§ 10^5')
    
    result = []

    sharps_count = 0

    # –∫–æ–ª-–≤–æ –∑–∞–∫—Ä—ã–≤–∞—é—â–∏—Ö —Å–∫–æ–±–æ–∫ –¥–ª—è —Å–æ–±–ª—é–¥–µ–Ω–∏—è –±–∞–ª–∞–Ω—Å–∞
    to_balance_count = 0

    # –∫–æ–ª-–≤–æ –∑–∞–∫—Ä—ã–≤–∞—é—â–∏—Ö —Å–∫–æ–±–æ–∫, –∫–æ—Ç–æ—Ä–æ–µ –º–æ–∂–Ω–æ –ø—Ä–∏–º–µ–Ω–∏—Ç—å –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã
    # —Å–æ–±–ª—é—Å—Ç–∏ —É—Å–ª–æ–≤–∏–µ: –¥–ª—è –∫–∞–∂–¥–æ–≥–æ i (1‚â§i‚â§|s|) —Å—Ä–µ–¥–∏ –ø–µ—Ä–≤—ã—Ö i —Å–∏–º–≤–æ–ª–æ–≤
    # —Å—Ç—Ä–æ–∫–∏ s —Å–∏–º–≤–æ–ª–æ–≤ ')' –Ω–µ –±–æ–ª—å—à–µ, —á–µ–º —Å–∏–º–≤–æ–ª–æ–≤ '('
    available_count = 0

    for symbol in src:
        if symbol == '(':
            to_balance_count += 1
        elif symbol == ')':
            to_balance_count -= 1
        elif symbol == '#':
            sharps_count += 1

    if to_balance_count < 1 or sharps_count < 1:
        return -1

    for symbol in src:
        if symbol == '(':
            available_count += 1
        elif symbol == ')':
            available_count -= 1

            if grab_closing(result):
                to_balance_count += 1
        elif symbol == '#':
            count = min(available_count, to_balance_count)

            if count < 1 and grab_closing(result):
                count += 1

            result.append(count)
            to_balance_count -= count
            sharps_count -= 1

            if sharps_count == 0:
                break

    for count in result:
        if count < 1:
            return -1

    return result


assert beautify_count('(((#)((#)') == [2, 1]
assert beautify_count('(((#((#') == [3, 2]
assert beautify_count('(#(((') == [1]
assert beautify_count('#(((') == -1
assert beautify_count('(((#') == [3]
assert beautify_count('(#(#(#') == [1, 1, 1]
assert beautify_count('(') == -1
assert beautify_count('((#)(((#') == [1, 3]
assert beautify_count('()((#((#(#()') == [2, 2, 1]

print('–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
src = input()

print('\n–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
print(beautify_count(src))
```

```python
def is_correct_sequence(sequence: str) -> bool:
    brackets_stack = []
    failed = False

    for char in sequence:
        if char == '(':
            brackets_stack.append('(')
        elif char == ')':
            if len(brackets_stack) > 0:
                brackets_stack.pop()
            else:
                failed = True
                break

    return not failed and len(brackets_stack) == 0


assert not is_correct_sequence('(')
assert not is_correct_sequence(')')
assert not is_correct_sequence('()()(')
assert not is_correct_sequence(')(')
assert is_correct_sequence('()')
assert is_correct_sequence('()()')
assert is_correct_sequence('(()())')


def correct_combinations_count(k: int, input_func):
    if not 1 <= k <= 3 * (10 ** 5):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 1 ‚â§ k ‚â§ 3 * 10^5')

    sequences = []
    lengths_sum = 0
    count = 0

    for i in range(k):
        sequences.append(input_func())
        lengths_sum += len(sequences[-1])

        if lengths_sum > 3 * (10 ** 5):
            raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ: —Å—É–º–º–∞ –¥–ª–∏–Ω –≤—Å–µ—Ö —Å–∫–æ–±–æ—á–Ω—ã—Ö –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 3‚ãÖ10^5')

        for j in range(len(sequences)):
            sequence = sequences[j]

            if is_correct_sequence(sequences[-1] + sequence):
                count += 1

            if i != j and is_correct_sequence(sequence + sequences[-1]):
                count += 1

    return count


print('–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
k = int(input())

count = correct_combinations_count(k, lambda: input())
print('\n–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
print(count)
```

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í—ã–≤–µ–¥–∏—Ç–µ –æ–¥–Ω–æ —Ü–µ–ª–æ–µ –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–µ–∫—É–Ω–¥, –ø–æ –∏—Å—Ç–µ—á–µ–Ω–∏–∏ –∫–æ—Ç–æ—Ä—ã—Ö –Ω–∞–≥—Ä—É–∑–∫–∞ —Å–µ—Ä–≤–µ—Ä–æ–≤ –±—É–¥–µ—Ç —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∞.

–ü—Ä–∏–º–µ—Ä—ã\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
2\
1 6\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
2

```python
def balance_duration(loads: list[int]) -> int:
    if not 1 <= len(loads) <= 10 ** 5:
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 1 ‚â§ –¥–ª–∏–Ω–∞ loads ‚â§ 10^5')

    for load in loads:
        if not 0 <= load <= 2 * (10 ** 4):
            raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 0 ‚â§ –∫–æ–ª-–≤–æ –∑–∞–ø—Ä–æ—Å–æ–≤ ‚â§ 2 ‚ãÖ 10^4')
    
    loads = sorted(loads)
    k = 0

    start = 0
    end = len(loads) - 1

    while start < end:
        while (loads[end] - 1) - (loads[start] + 1) >= 0:
            loads[start] += 1
            loads[end] -= 1
            k += 1

        start += 1
        end -= 1

    return k


assert balance_duration([1, 6]) == 2
assert balance_duration([10, 11, 10, 11, 10, 11, 11]) == 0
assert balance_duration([1, 2, 3, 4, 5]) == 3

print('–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
loads = list(map(int, input().split()))

print('\n–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
print(balance_duration(loads))
```

–í–∞–º –∏–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –∑–∞–º–æ–∫ –æ—Ç–∫—Ä–æ–µ—Ç—Å—è, –µ—Å–ª–∏ –Ω–∞ –¥–∏—Å–ø–ª–µ–µ –±—É–¥–µ—Ç –æ—Ç–æ–±—Ä–∞–∂–∞—Ç—å—Å—è –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –∏–∑ –≤—Å–µ—Ö, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å –Ω–∞–∂–∞—Ç–∏–µ–º –Ω–∞ –∫–Ω–æ–ø–∫–∏ –≤ –∫–∞–∫–æ–º-—Ç–æ –ø–æ—Ä—è–¥–∫–µ (–∫–∞–∂–¥–∞—è –∫–Ω–æ–ø–∫–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å –Ω–∞–∂–∞—Ç—å –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ —Ä–∞–∑). –ü—Ä–∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏–∏ —á–∏—Å–µ–ª –≤–µ–¥—É—â–∏–µ –Ω—É–ª–∏ –∏–≥–Ω–æ—Ä–∏—Ä—É—é—Ç—Å—è. –¢–µ–ø–µ—Ä—å –≤–∞–º —Ç—Ä–µ–±—É–µ—Ç—Å—è –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, —á—Ç–æ —ç—Ç–æ –∑–∞ —á–∏—Å–ª–æ.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ n (1‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ1000) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–∏—Ñ—Ä –Ω–∞ –¥–∏—Å–ø–ª–µ–µ.

–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ n —Ü–∏—Ñ—Ä ‚Äî –∏—Å—Ö–æ–¥–Ω–æ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ –¥–∏—Å–ø–ª–µ—è.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í—ã–≤–µ–¥–∏—Ç–µ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É –∏–∑ n —Ü–∏—Ñ—Ä ‚Äî –∏—Å–∫–æ–º–æ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ –¥–∏—Å–ø–ª–µ—è, —Å–æ–¥–µ—Ä–∂–∞—â–µ–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–µ —á–∏—Å–ª–æ.

–ü—Ä–∏–º–µ—Ä—ã\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
3\
579\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
024

```python
def min_code(code: str) -> str:
    if not 1 <= len(code) <= 1000:
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 1 ‚â§ –¥–ª–∏–Ω–∞ code ‚â§ 1000')
    
    digits = [int(digit) for digit in code]
    n = len(digits)

    min_index = 0
    min_digit = digits[0]

    for i in range(1, n):
        digit = digits[i]
        if digit < min_digit:
            min_digit = digit
            min_index = i

    shifted_digits = [0] * n
    shift = n - min_index

    for i in range(n):
        digit = digits[i]
        shifted_digits[(i + shift) % n] = digit

    if shifted_digits[0] > 0:
        add_count = 10 - shifted_digits[0]

        for i in range(n):
            shifted_digits[i] = (shifted_digits[i] + add_count) % 10

    return ''.join(map(str, shifted_digits))


assert min_code('579') == '024'
assert min_code('2014') == '0142'
assert min_code('328') == '061'

print('–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
code = input()

print('\n–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
print(min_code(code))
```

```python
def find_by_first(data: list[tuple[int, int]], first: int) -> int:
    for pair in data:
        if pair[0] == first:
            return pair[1]

    return -1


def ordered_queue(k: int, students_neighbors: list[tuple[int, int]]) -> list[int]:
    result = [0] * k
    first = None

    if not 2 <= k <= 2 * (10 ** 5):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 2 ‚â§ k ‚â§ 2 ‚ãÖ 10^5')

    for neighbors in students_neighbors:
        for number in neighbors:
            if not 0 <= number <= 10 ** 6:
                raise ValueError('–Ω–∞—Ä—É—à–µ–Ω–æ —É—Å–ª–æ–≤–∏–µ 0 ‚â§ –Ω–æ–º–µ—Ä —Å—Ç—É–¥ –±–∏–ª–µ—Ç–∞ ‚â§ 10^6')

    # –ù–∞—Ö–æ–¥–∏–º –ø–µ—Ä–≤–æ–≥–æ —Å—Ç—É–¥–µ–Ω—Ç–∞ –≤ –æ—á–µ—Ä–µ–¥–∏
    for i in range(len(students_neighbors)):
        found = True

        for j in range(len(students_neighbors)):
            if i != j and students_neighbors[i][0] == students_neighbors[j][1]:
                found = False
                break

        if found:
            first = students_neighbors[i][0]
            break

    # –¥–∞–ª–µ–µ, –∑–Ω–∞—è –ø–µ—Ä–≤–æ–≥–æ —Å—Ç—É–¥–µ–Ω—Ç–∞ –∏ —Ç–æ, —á—Ç–æ –ø–µ—Ä–µ–¥ –Ω–∏–º —Å—Ç–æ–∏—Ç –Ω–æ–ª—å,
    # –º–æ–∂–Ω–æ –ª–µ–≥–∫–æ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –æ—á–µ—Ä–µ–¥—å, –Ω–∞–ø—Ä–∏–º–µ—Ä:
    # –≤ –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏–º–µ—é—Ç—Å—è –ø–∞—Ä—ã (0, 7) –∏ (92, 15), –ø—Ä–∏ —ç—Ç–æ–º
    # –ø–µ—Ä–≤—ã–º —ç–ª–µ–º–µ–Ω—Ç–æ–º —è–≤–ª—è–µ—Ç—Å—è —á–∏—Å–ª–æ 92. –û—Ç—Å—é–¥–∞ —Å–ª–µ–¥—É–µ—Ç, —á—Ç–æ —á–∏—Å–ª–æ 7 —Å—Ç–æ–∏—Ç
    # –ø–æ—Å–ª–µ 92 (–æ—Ç –Ω—É–ª—è —Å–¥–µ–ª–∞–≤ –¥–≤–∞ —à–∞–≥–∞ –≤–ø—Ä–∞–≤–æ), –∞ —á–∏—Å–ª–æ 15 –ø–æ—Å–ª–µ 7 (–æ—Ç 92 –¥–≤–∞ —à–∞–≥–∞ –≤–ø—Ä–∞–≤–æ).
    # –ü–æ–ª—É—á–∞–µ–º: 92, 7, 15
    result[0] = first
    for i in range(0, k - 1, 2):
        result[i + 1] = find_by_first(students_neighbors, 0 if i - 1 < 0 else result[i - 1])

        if i + 2 < len(result):
            result[i + 2] = find_by_first(students_neighbors, result[i])

    return result


assert ordered_queue(
    8,
    [(17, 0), (31, 20), (141, 17), (20, 2), (7, 31), (15, 141), (0, 7), (92, 15)]
) == [92, 7, 15, 31, 141, 20, 17, 2]

assert ordered_queue(
    4,
    [(92, 31), (0, 7), (31, 0), (7, 141)]
) == [92, 7, 31, 141]


print('–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
k = int(input())

students_neighbors = []

for i in range(k):
    a, b = map(int, input().split())
    students_neighbors.append((a, b))

print('\n–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
print(ordered_queue(k, students_neighbors))
```
